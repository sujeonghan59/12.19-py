```python
import requests  import webdriver # 브라우저 연결
from bs4 import BeautifulSoup as bs #  인터넷으로 긁어온 데이터의 정체
import pandas as pd

song_data = []
rank = 1

# 전체 노래 100곡 다시 추출

driver = webdriver.Chrome('chromedriver.exe')  # 드라이버 위치
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)

html = driver.page_source

soup = bs(html)
songs = soup.select('tbody > tr')


# 100번 반복하여 전체 노래 제목, 가수명 모두 추출
for i in range(0,100):
    song = songs[i]
    title = song.select('div.rank01 > span > a')[0].text
    singer = song.select(''div.rank01 > span > a')[0].text
    song_data.append(['Melon', rank, title, singer])
    rank += 1

    # 2중 리스트로 만든 100곡을 데이터프레임으로 전환하기

    df = pd.DataFrame(song_data, columns = ['서비스', '순위', '노래제목', '가수명'])
    df   # 판다스 데이터 프레인 가로 세로 형태로  
         # 데이터 프레인을 엑셀 파일로 저장하기

    df.to_excel('Melon.xlsx', index=False)
```
